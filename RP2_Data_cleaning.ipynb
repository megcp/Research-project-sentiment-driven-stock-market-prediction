## Cleaning Stock data

```python
stocks = pd.read_csv('/content/drive/My Drive/Research Project/apple_stock.csv')
```

```python
# convert 'Time Published' into datetime
stocks['timestamp'] = pd.to_datetime(stocks['timestamp'])
```

```python
stocks
```

```python
# remove data that exceed artciles
cut_of_date = pd.to_datetime('2022-03-07 00:00:00')
stocks = stocks[stocks['timestamp'] >= cut_of_date]
```

```python
stocks
```

```python
stocks= stocks.drop(columns=['Unnamed: 0'])
```

```python
stocks
```

```python
# remove stock for the 2024-06-07
stocks = stocks.drop(stocks.index[0])
```

```python
stocks
```

```python
# checks for duplicates
duplicates = stocks.duplicated()
print(duplicates.sum())
```

```python
stocks.to_csv('/content/drive/My Drive/Research Project/clean_apple_stocks_2022_2024.csv', index=False)
```
